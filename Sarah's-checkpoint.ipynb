```python
def remove_outliers_iqr(df, column_name, threshold=1.5):
    Q1 = bankDF[column_name].quantile(0.03) # Calculate the first quartile (Q1) at 3rd percentile
    Q3 = bankDF[column_name].quantile(0.97) # Calculate the third quartile (Q3) at 97th percentile
    IQR = Q3 - Q1 # Calculate the Interquartile Range (IQR)
    lower_bound = Q1 - threshold * IQR # Calculates the lower bound for the outliers
    upper_bound = Q3 + threshold * IQR # Calculates the upper bound for the outliers
    return bankDF[(bankDF[column_name] >= lower_bound) & (bankDF[column_name] <= upper_bound)]
# Return a DataFrame with outliers removed based on the specified column and bounds

# The columns on which i want to detect outliers for.
columns_to_check = [ "age", "duration", "campaign", "pdays", "previous",
    "emp.var.rate", "cons.price.idx", "cons.conf.idx", "euribor3m", "nr.employed",
    "job_Label", "marital_Label", "education_Label", "default_Label",
    "housing_Label", "loan_Label", "contact_Label", "month_Label",
    "day_of_week_Label", "poutcome_Label", "y_Label"]

# Apply the remove_outliers_iqr function to detect and remove outliers
data_cleaned = bankDF.copy() # Create a copy of the original DataFrame 'bankDF' to store the cleaned data.
for column in columns_to_check: # Loop through the specified columns and apply the 'remove_outliers_iqr' function
    data_cleaned = remove_outliers_iqr(data_cleaned, column) # Remove outliers from the current column using the 'remove_outliers_iqr' function
    
# Update the original 'bankDF' with the cleaned data from 'data_cleaned'
bankDF = data_cleaned
```

```python


# Define a function to plot the learning curve for training and testing data
def plot_learning_curve(modelTy, title, X, Y, ylim=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure() # Create a new figure for the learning curve plot
    plt.title(title)
    if ylim is not None: 
        plt.ylim(*ylim) # Set the y-axis limits, if specified

    plt.xlabel("Training Samples")
    plt.ylabel("Accuracy")

    # Generate learning curves using the learning_curve function
    train_sizes, train_scores, test_scores = learning_curve(
        modelTy, X, Y, train_sizes=train_sizes, cv=ShuffleSplit(n_splits=5, test_size=0.2, random_state=42), n_jobs=None)

    # Calculate mean and standard deviation of training and testing scores
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.grid()
    
    # Fill the area between the training score curves

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    # Fill the area between the testing score curves
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    
    # Plot the training score curve
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training Accuracy")
    
    # Plot the testing score curve
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Testing Accuracy")

    plt.legend(loc="best")
    return plt

# call the function
plot_learning_curve(tree, "Learning Curve (Decision Tree)", X, Y)

plt.show()
```

```python
# Define a function to plot the learning curve for training and testing data
def plot_learning_curve(estimator, title, X, y, ylim=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure() # Create a new figure for the learning curve plot.
    plt.title(title)  # Set the title of the plot
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel(" Samples")
    plt.ylabel("Accuracy")

    # Generate learning curves using the learning_curve function
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, Y, train_sizes=train_sizes, cv=ShuffleSplit(n_splits=5, test_size=0.2, random_state=42), n_jobs=None)

    # Calculate mean and standard deviation of training and testing scores
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.grid()  # Add a grid to the plot

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r") # Fill the area between the training score curves
    
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")  # same but for testing
    
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training Accuracy")  # plot training
    
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Testing Accuracy") #plot testing

    plt.legend(loc="best")                  #add legend
    return plt

# calling the function but with my Gradient Boosting Classifier
plot_learning_curve(grad, "Learning Curve (Gradient Boosting)", X, Y)

plt.show()
```

```python
# Define a function to plot the learning curve for training and testing data
def plot_learning_curve(estimator, title, X, Y, ylim=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure() #create a new fig
    plt.title(title) # set the title
    if ylim is not None:      #set limit if not already 
        plt.ylim(*ylim)
    plt.xlabel("Training samples")
    plt.ylabel("accuracy")

    # calculate training and testing scores using learning curve
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, Y, train_sizes=train_sizes, cv=ShuffleSplit(n_splits=5, test_size=0.2, random_state=42), n_jobs=None)

    # Calculate the training and testing scores mean and standard deviation
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.grid() #add grid lines lel plot

     # Fill the area around the training score line with color for variance purpose 
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    
     # Fill the area around the test score line with color for variance purpose 
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    #plot training
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training Accuracy")
    
    #plot testing
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Testing Accuracy")

    plt.legend(loc="best")
    return plt

#calling the function with Knn Classifier
plot_learning_curve(knn, "Learning Curve (KNN)", X, Y)

plt.show()
```
